```python
import pandas as pd
import yfinance as yf
import numpy as np
import plotly.graph_objects as go
```

```python

# 保值因子 https://bigquant.com/wiki/doc/zysspc4V3G
tips = "TIP" # 美國10年期抗通脹債券ETF -> 美國實際利率 ma40 > ma40.shift(1) -> short; long elsewhere
# 美國10年期國債利率 - 通脹保護債券 -> 美國通脹預期 > 2.2% -> Long
bonds_etf = 'IEF' # iShares 7-10年期美國公債ETF
gold_etf = 'IAU'
tip = yf.download(tips, progress=False)
ief = yf.download(bonds_etf, progress=False)
gold = yf.download(gold_etf, progress=False)
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(
    rows=1, cols=1,  
    shared_xaxes=True,
    vertical_spacing=0.1,  
    specs=[[{"secondary_y": True}]]  
)

# Cumulative Profit Line (Primary y-axis)
fig.add_trace(go.Scatter(
    x=ief.index, y=ief['Close'],
    mode="lines", name="tip", line=dict(color="blue")
), row=1, col=1, secondary_y=False)  

# OP Line (Secondary y-axis)
fig.add_trace(go.Scatter(
    x=gold.index, y=gold['Close'],
    mode="lines", name="gold", line=dict(color="orange")
), row=1, col=1, secondary_y=True)  

fig.show()
```
